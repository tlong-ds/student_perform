# Basic Tests for Relationship with Exam_Score


```python
# Basic Tests for Relationship with Exam_Score

# T-test for 'Exam_Score' by 'Gender'
male_scores = df[df['Gender'] == 'Male']['Exam_Score']
female_scores = df[df['Gender'] == 'Female']['Exam_Score']
t_stat, p_value = stats.ttest_ind(male_scores, female_scores)
print(f"T-test for Exam Score by Gender: t-statistic = {t_stat:.2f}, p-value = {p_value:.2f}")

# Chi-square test for 'Exam_Score' by 'School_Type'
# Binning 'Exam_Score' for chi-square test
df['Exam_Score_Binned'] = pd.cut(df['Exam_Score'], bins=5, labels=False)
contingency_table = pd.crosstab(df['Exam_Score_Binned'], df['School_Type'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-square test for Exam Score by School Type: chi2 = {chi2:.2f}, p-value = {p:.2f}")

# Chi-square test for 'Exam_Score' by 'Parental_Involvement'
contingency_table_parental = pd.crosstab(df['Exam_Score_Binned'], df['Parental_Involvement'])
chi2_parental, p_parental, dof_parental, expected_parental = stats.chi2_contingency(contingency_table_parental)
print(f"Chi-square test for Exam Score by Parental Involvement: chi2 = {chi2_parental:.2f}, p-value = {p_parental:.2f}")

# T-test for 'Exam_Score' by 'Internet_Access'
internet_yes_scores = df[df['Internet_Access'] == 'Yes']['Exam_Score']
internet_no_scores = df[df['Internet_Access'] == 'No']['Exam_Score']
t_stat_internet, p_value_internet = stats.ttest_ind(internet_yes_scores, internet_no_scores)
print(f"T-test for Exam Score by Internet Access: t-statistic = {t_stat_internet:.2f}, p-value = {p_value_internet:.2f}")

# T-test for 'Exam_Score' by 'Tutoring_Sessions'
tutoring_yes_scores = df[df['Tutoring_Sessions'] > 0]['Exam_Score']
tutoring_no_scores = df[df['Tutoring_Sessions'] == 0]['Exam_Score']
t_stat_tutoring, p_value_tutoring = stats.ttest_ind(tutoring_yes_scores, tutoring_no_scores)
print(f"T-test for Exam Score by Tutoring Sessions: t-statistic = {t_stat_tutoring:.2f}, p-value = {p_value_tutoring:.2f}")
```

# T-Test
Perform a t-test to analyze the relationship between Exam_Score and other variables.

```python
# T-Test

# T-test for 'Exam_Score' by 'Gender'
male_scores = df[df['Gender'] == 'Male']['Exam_Score']
female_scores = df[df['Gender'] == 'Female']['Exam_Score']
t_stat_gender, p_value_gender = stats.ttest_ind(male_scores, female_scores)
print(f"T-test for Exam Score by Gender: t-statistic = {t_stat_gender:.2f}, p-value = {p_value_gender:.2f}")

# T-test for 'Exam_Score' by 'School_Type'
public_scores = df[df['School_Type'] == 'Public']['Exam_Score']
private_scores = df[df['School_Type'] == 'Private']['Exam_Score']
t_stat_school, p_value_school = stats.ttest_ind(public_scores, private_scores)
print(f"T-test for Exam Score by School Type: t-statistic = {t_stat_school:.2f}, p-value = {p_value_school:.2f}")

# T-test for 'Exam_Score' by 'Internet_Access'
internet_yes_scores = df[df['Internet_Access'] == 'Yes']['Exam_Score']
internet_no_scores = df[df['Internet_Access'] == 'No']['Exam_Score']
t_stat_internet, p_value_internet = stats.ttest_ind(internet_yes_scores, internet_no_scores)
print(f"T-test for Exam Score by Internet Access: t-statistic = {t_stat_internet:.2f}, p-value = {p_value_internet:.2f}")

# T-test for 'Exam_Score' by 'Parental_Involvement'
low_parental_involvement_scores = df[df['Parental_Involvement'] == 'Low']['Exam_Score']
medium_parental_involvement_scores = df[df['Parental_Involvement'] == 'Medium']['Exam_Score']
high_parental_involvement_scores = df[df['Parental_Involvement'] == 'High']['Exam_Score']

t_stat_parental_low_medium, p_value_parental_low_medium = stats.ttest_ind(low_parental_involvement_scores, medium_parental_involvement_scores)
t_stat_parental_low_high, p_value_parental_low_high = stats.ttest_ind(low_parental_involvement_scores, high_parental_involvement_scores)
t_stat_parental_medium_high, p_value_parental_medium_high = stats.ttest_ind(medium_parental_involvement_scores, high_parental_involvement_scores)

print(f"T-test for Exam Score by Parental Involvement (Low vs Medium): t-statistic = {t_stat_parental_low_medium:.2f}, p-value = {p_value_parental_low_medium:.2f}")
print(f"T-test for Exam Score by Parental Involvement (Low vs High): t-statistic = {t_stat_parental_low_high:.2f}, p-value = {p_value_parental_low_high:.2f}")
print(f"T-test for Exam Score by Parental Involvement (Medium vs High): t-statistic = {t_stat_parental_medium_high:.2f}, p-value = {p_value_parental_medium_high:.2f}")

# T-test for 'Exam_Score' by 'Tutoring_Sessions'
tutoring_yes_scores = df[df['Tutoring_Sessions'] > 0]['Exam_Score']
tutoring_no_scores = df[df['Tutoring_Sessions'] == 0]['Exam_Score']
t_stat_tutoring, p_value_tutoring = stats.ttest_ind(tutoring_yes_scores, tutoring_no_scores)
print(f"T-test for Exam Score by Tutoring Sessions: t-statistic = {t_stat_tutoring:.2f}, p-value = {p_value_tutoring:.2f}")
```

# Chi-Square Test
Perform a chi-square test to analyze the relationship between Exam_Score and other variables.

```python
# Chi-Square Test

# Chi-square test for 'Exam_Score' by 'School_Type'
# Binning 'Exam_Score' for chi-square test
df['Exam_Score_Binned'] = pd.cut(df['Exam_Score'], bins=5, labels=False)
contingency_table_school = pd.crosstab(df['Exam_Score_Binned'], df['School_Type'])
chi2_school, p_school, dof_school, expected_school = stats.chi2_contingency(contingency_table_school)
print(f"Chi-square test for Exam Score by School Type: chi2 = {chi2_school:.2f}, p-value = {p_school:.2f}")

# Chi-square test for 'Exam_Score' by 'Parental_Involvement'
contingency_table_parental = pd.crosstab(df['Exam_Score_Binned'], df['Parental_Involvement'])
chi2_parental, p_parental, dof_parental, expected_parental = stats.chi2_contingency(contingency_table_parental)
print(f"Chi-square test for Exam Score by Parental Involvement: chi2 = {chi2_parental:.2f}, p-value = {p_parental:.2f}")

# Chi-square test for 'Exam_Score' by 'Gender'
contingency_table_gender = pd.crosstab(df['Exam_Score_Binned'], df['Gender'])
chi2_gender, p_gender, dof_gender, expected_gender = stats.chi2_contingency(contingency_table_gender)
print(f"Chi-square test for Exam Score by Gender: chi2 = {chi2_gender:.2f}, p-value = {p_gender:.2f}")

# Chi-square test for 'Exam_Score' by 'Internet_Access'
contingency_table_internet = pd.crosstab(df['Exam_Score_Binned'], df['Internet_Access'])
chi2_internet, p_internet, dof_internet, expected_internet = stats.chi2_contingency(contingency_table_internet)
print(f"Chi-square test for Exam Score by Internet Access: chi2 = {chi2_internet:.2f}, p-value = {p_internet:.2f}")

# Chi-square test for 'Exam_Score' by 'Tutoring_Sessions'
contingency_table_tutoring = pd.crosstab(df['Exam_Score_Binned'], df['Tutoring_Sessions'])
chi2_tutoring, p_tutoring, dof_tutoring, expected_tutoring = stats.chi2_contingency(contingency_table_tutoring)
print(f"Chi-square test for Exam Score by Tutoring Sessions: chi2 = {chi2_tutoring:.2f}, p-value = {p_tutoring:.2f}")
```
